### <span style="color:blue"> Brief Introduction to deep learning-
Deep learning is a branch/division of Machine Learning where the models are a chain of geometric fucntions, applied one after other. Operations are structured into modules called layers. A deep learning module is stack of layers, more specifically a graph of layers.

These layers are parametarized by weights which are learned during the training. The ***Knowledge*** of a model is stored in its weights which are parameters learned during the training. The process of learing is intented to find good values of these weights. 

### Importing a few basic libraries for data handling: 

**numpy:** This library is added for array operation. It is much faster than arrays/ lists and offer most linear algebra operation 

pandas: This library is added for data manipulation and analysis, It helps to us to read various file and data processing. In simpler terms it offers excel/spreadsheet like features in Phythonic interface. 

```python
import numpy as np 
import pandas as pd
```

### Reading the data provided in the competetion. 
Pandas offers a read_csv function that allows to read the comma seprated values to a pandas data frame and allows the user to study and analyse the data to find trends and relationship in data. 

The drop function will drop a particular row/column depending on the axis specified in the method argument. 

```python
training_data = pd.read_csv("C:\Users\Ashish Papanai\Downloads\Digit Recogniser (First NN)\data\train.csv")
test_images = pd.read_csv("C:\Users\Ashish Papanai\Downloads\Digit Recogniser (First NN)\data\test.csv")

train_images = training_data.drop("label", axis=1)
train_labels = training_data["label"]
```

Neural Network take data as input in form of Numpy Arrays, the data frame thus needs to be converted to numpy arrays so that it can become ready for processing by the neural network. 

The to_numpy() function helps us to convert to numpy array. 

```python
train_images.to_numpy()
train_labels.to_numpy()
test_images.to_numpy()
```

**Data Preprocessing:**  The data values are divided by 255 for scaling the values in the numpy arrays, this will make the values of array elements in between 0-1 which was previously in between 0-255

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

The labels are also of same dimensions as the training data, the training labels needs to be converted categorically to a Original_Dimension * Class dimensional array, this will help the model to understand which class does the data belongs to. 

In simpler terms it will categorise the data. 

```python
from keras.utils import to_categorical

train_labels = to_categorical(train_labels)
```

### <span style="color:blue"> Model Generation using Keras with TensorFlow Backend: </span>
A black box processing the data and genrating geometrical relationship between the data and label is called a model in deep learning. 

To create models in keras we import **models** class from the keras library. The layers of models are created by various functions **layers** class, the functions describe the type of layer we want to create in the model. 

```python
from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28, )))
network.add(layers.Dense(10, activation='softmax'))
network.summary()
```

### Compiling the network:
The model is compiled with RMSprop optimiser, which works similar to Gradient Descent, The Formula for RMSprop optimser is:



These optimisers are available inside keras optimisers. 

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

#### Final Step is to fit the model to the training images and the training labels and the model tries that for n numbers of epochs taking batch size number of training_images in one iteration. 

***Epochs***: Number of iterations through the complete training data. 

***Batch Size***: Number of training samples taken in one iteration of an epoch. 

```python
network.fit(train_images, train_labels, epochs = 5, batch_size=128)
```

### <span style="color:RED">The following cells are used to submit and prepare the submission.csv </span>
We need to prepare a csv file with the labels predicted for the test data provided to us in the competition. A sample submission file is always attached to the data set provided by the competition hosts. 

```python
submit = pd.DataFrame(np.argmax(network.predict(test_images), axis=1), 
                      columns=['Label'], 
                      index=pd.read_csv('../input/digit-recognizer/sample_submission.csv')['ImageId'])

submit.index.name = 'ImageId'
submit.to_csv('submittion.csv')
```
